```python
import numpy as np
import tensorflow as tf
tf.reset_default_graph()
from keras.preprocessing.image import ImageDataGenerator
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("https://i.imgur.com/jZqpV51.png")
img = np.array(img, dtype=float) / 255.0

def display_image(img):
    plt.figure(); plt.imshow(img)
    plt.grid(False);  plt.axis('off'); plt.show()

display_image(img)
```

### The image above, taken from the dataset (https://www.kaggle.com/paultimothymooney/chest-xray-pneumonia/home), shows X-Ray images of a normal chest, a bacterial Pneumonia, and finally a viral Pneumonia. Using convolutional neural networks, the task is to classify images as normal or infected. 

```python
rescale = 1./255
IMG_HEIGHT = 224
IMG_WIDTH = 224
MINI_BATCH = 32
PATH = "/Users/hyunjaecho/Desktop/chest_xray/"

train_generator = ImageDataGenerator(rescale=rescale).flow_from_directory(
    PATH + "train",
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=1,
    class_mode='binary')

# Since the validation set is too small, I use the test set to validate my answers.
# I do not use the validation set.
v_generator = ImageDataGenerator(rescale=rescale).flow_from_directory(
    PATH + "test",
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=1,
    class_mode='binary')
```

```python
# Variables
NUM_CHANNELS = 3
NUM_LABELS = 2
LEARNING_RATE = 0.01

with tf.name_scope("hidden-layer"):
    conv1_weights = tf.Variable(tf.truncated_normal([7, 7, NUM_CHANNELS, MINI_BATCH]))
    conv1_bias = tf.Variable(tf.zeros([MINI_BATCH]))

    conv2_weights = tf.Variable(tf.truncated_normal([7, 7, 32, 64]))
    conv2_bias = tf.Variable(tf.zeros([64]))

    fc1_weights = tf.Variable(tf.truncated_normal([int(np.ceil(IMG_HEIGHT / 4)) * int(np.ceil(IMG_WIDTH / 4)) * 64, 512]))
    fc1_biases = tf.Variable(tf.constant(0.1, shape=[512]))

    fc2_weights = tf.Variable(tf.truncated_normal([512, NUM_LABELS]))
    fc2_biases = tf.Variable(tf.constant(0.1, shape=[NUM_LABELS]))
```

```python
def model(data):
    # conv layer 1
    print("DATA: ", data.shape)
    conv = tf.nn.conv2d(data, filter=conv1_weights, strides=[1,1,1,1], padding='SAME')
    conv = tf.nn.bias_add(conv, conv1_bias)
    activate = tf.nn.relu(conv)
    dropout = tf.nn.dropout(activate, 0.5)
    pool = tf.nn.max_pool(dropout, ksize=[1,2,2,1], strides=[1, 2, 2, 1], padding='SAME')

    # conv layer 2
    conv = tf.nn.conv2d(pool, filter=conv2_weights, strides=[1,1,1,1], padding='SAME')
    conv = tf.nn.bias_add(conv, conv2_bias)
    activate = tf.nn.relu(conv)
    dropout = tf.nn.dropout(activate, 0.5)
    
    pool = tf.nn.max_pool(activate, ksize=[1,2,2,1], strides=[1, 2, 2, 1], padding='SAME')
    pool_shape = pool.get_shape().as_list()
    pool = tf.convert_to_tensor(pool)
    
    reshape = tf.reshape(pool, [pool_shape[0], pool_shape[1] * pool_shape[2] * pool_shape[3]])
    
    # fc layer 1
    fc1 = tf.matmul(reshape, fc1_weights) + fc1_biases
    fc1_activate = tf.nn.relu(fc1)
    fc1 = tf.nn.dropout(fc1_activate, 0.5)
    
    # fc layer 2
    fc2 = tf.matmul(fc1, fc2_weights) + fc2_biases
    
    return fc2
```

```python
# placeholders
with tf.name_scope("placeholders"):
    train_batch = tf.placeholder(dtype=tf.float32, shape=[1, IMG_HEIGHT, IMG_WIDTH, NUM_CHANNELS])
    train_labels = tf.placeholder(dtype=tf.int64, shape=[2,])
    
    eval_batch = tf.placeholder(dtype=tf.float32, shape=[1, IMG_HEIGHT, IMG_WIDTH, NUM_CHANNELS])
    eval_labels = tf.placeholder(dtype=tf.int64, shape=[2,])
```

```python
logits = model(train_batch)

# Predictions for the current training minibatch.
train_prediction = tf.nn.softmax(logits)

# Validation prediction
eval_prediction = tf.nn.softmax(model(eval_batch))

# We make sure that the weights and the biases do not grow too much by including the l2 loss of them into the loss function.
with tf.name_scope("loss"):
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=train_labels, logits=logits))
    regularizers = (tf.nn.l2_loss(fc1_weights)
                + tf.nn.l2_loss(fc1_biases)
                + tf.nn.l2_loss(fc2_weights)
                + tf.nn.l2_loss(fc2_biases))
    
    alpha = 5e-2
    loss += alpha * regularizers
    
with tf.name_scope("optimizer"):
    optimizer = tf.train.AdamOptimizer(learning_rate=LEARNING_RATE).minimize(loss)
```

```python
def recall(true_positive, false_negative):
    return true_positive/(true_positive + false_negative)

def precision(true_positive, false_positive):
    return true_positive/(true_positive + false_positive)

true_positive = 0
false_negative = 0
false_positive = 0
true_negative = 0
f1s = []

def compute_f1(predictions, labels):
    global true_positive, false_negative, false_positive, true_negative
    global f1s
    
    for i, pred in enumerate(predictions):
        pos_of_1 = pred.index(1)
        label_pos_of_1 = labels[i].index(1)
        
        if pos_of_1 == 1 and label_pos_of_1 == 1:
            true_positive += 1
        elif pos_of_1 == 0 and label_pos_of_1 == 1:
            false_negative += 1
        elif pos_of_1 == 1 and label_pos_of_1 == 0:
            false_positive += 1
        elif pos_of_1 == 0 and label_pos_of_1 == 0:
            true_negative += 1
            
    print("true_positive: {}, false_negative: {}, false_positive: {}, true_negative: {}".format(true_positive, false_negative, false_positive, true_negative))
    
    p = precision(true_positive, false_positive)
    r = recall(true_positive, false_negative)
    print("precision: {}, recall: {}".format(p, r))
    f1 = 2*p*r/(p+r)
    
    f1s.append(f1)
    
    return f1
```

```python
# Training
"""
NORMAL = 0
PNEUMONIA = 1

train_generator: 163 of minibatches of training data
train_generator[i] = 32 images
train_generator[i][j] = i,j -th image, shape (IMG_HEIGHT, IMG_WIDTH, CHANNELS=3)

"""
EPOCHS = 1

training_loss = []

with tf.Session() as sess:
    print("Session started")
    sess.run(tf.global_variables_initializer())
    
    with tf.name_scope("summaries"):
        tf.summary.scalar("loss", loss)
        writer = tf.summary.FileWriter("./nn_logs", sess.graph)
        merged = tf.summary.merge_all()
    
    
    for epoch in range(EPOCHS):
#         # Training
        for i in range(train_generator.samples):
            train_data, train_label = train_generator[i][0], train_generator[i][1]
            # one hot encode label
            if train_label == 1:
                train_label = [0, 1]
            else:
                train_label = [1, 0]

            feed_dict = {train_batch:train_data, train_labels:train_label}

            sess.run(optimizer, feed_dict=feed_dict)

            if i % 100 == 0:
                print(i, "-th iteration of training")
                l, predictions = sess.run([loss, train_prediction], feed_dict=feed_dict)
                training_loss.append(l)
                print("training_loss: {}".format(l))

        # Validation
        k = 0
        val_predictions = []
        val_labels = []
        
        for k in range(v_generator.samples):
            val_data, val_label = v_generator[k][0], v_generator[k][1]
            
            # one hot encode label
            if val_label == 1:
                val_label = [0, 1]
            else:
                val_label = [1, 0]
                
            val_feed_dict = {eval_batch:val_data, eval_labels:val_label}
            val_prediction = sess.run(eval_prediction, feed_dict=val_feed_dict)
            
            val_prediction = val_prediction[0]
            
            if val_prediction[0] < val_prediction[1]:
                val_prediction = [0, 1]
            else:
                val_prediction = [1, 0]
                
            val_predictions.append(val_prediction)
            val_labels.append(val_label)
            
#             compute_f1(val_predictions, val_labels)
            
            if k % 10 == 0 and k > 0:
                print(k, "-th iteration of validation")
                
            if k == v_generator.samples - 1:
                f1 = compute_f1(val_predictions, val_labels)
                print("validation F1 score: {}".format(f1))
                
            k += 1
```

Since the distribution between pneumonia and normal is highly unbalanced, calculating the F1 score is more reliable than (number of right predictions / total number of predictions) measurement. F1 score of 1 is considered perfect and 0 is considered totally wrong.

```python
plt.title('training_loss progress')
plt.ylabel('loss')
plt.xlabel('iteration #')
plt.plot(training_loss)
plt.show()
```

We can see that the training loss goes down as training progresses. Further training will improve the model.

We notice that the precision, the recall, and the F1 score are around 0.68, 0.84 and 0.75 respectively. There can be a few wats for improving these scores. First, the model was too simple. I used a simple 2-layered convolutional neural networks, while a more complicated, and perhaps pretrained, models would have performed better. Second, the training time was not long enough. Third, the input data was too small. Using a pretrained model can help overcome the first two potential reasons.

```python
import matplotlib
matplotlib.rc('image', cmap = 'gray')
from keras.models import Model
from keras.layers import Dense, Dropout, Flatten
from sklearn.metrics import f1_score
from keras.layers import BatchNormalization

from keras.applications.vgg19 import VGG19

VGG19 = VGG19(weights = None, include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))

vgg19_output = VGG19.output
# vgg19_output = Dropout(0.5)(vgg19_output)
vgg19_output = Flatten()(vgg19_output)
vgg19_output = Dense(128, activation='relu')(vgg19_output)
# vgg19_output = BatchNormalization()(vgg19_output)
vgg19_predictions = Dense(2, activation='softmax')(vgg19_output)
model = Model(inputs=VGG19.input, outputs=vgg19_predictions)

# First: train only the top layers (which were randomly initialized)
for layer in VGG19.layers:
    layer.trainable = False
    
model.compile(loss='categorical_crossentropy', 
              optimizer='adam')

model.summary()
```

```python
training_data = []
training_labels = []
for i in range(train_generator.samples):
        batch_data, batch_label = train_generator[i][0], train_generator[i][1]
        batch_data = batch_data.reshape(IMG_HEIGHT, IMG_WIDTH, 3)
        training_data.append(batch_data)
        if batch_label == 0:
            training_labels.append([1, 0])
        else:
            training_labels.append([0, 1])
        
training_data, training_labels = np.array(training_data), np.array(training_labels)
```

```python

model.fit(training_data, training_labels, epochs=1, batch_size=MINI_BATCH, verbose=1)
```

```python
validation_data = []
validation_labels = []
for k in range(v_generator.samples):
    val_data, val_label = v_generator[k][0], v_generator[k][1]
    val_data = val_data.reshape(IMG_HEIGHT, IMG_WIDTH, 3)
    validation_data.append(val_data)
    if val_label == 0:
        validation_labels.append([1, 0])
    else:
        validation_labels.append([0, 1])
        
validation_data, validation_labels = np.array(validation_data), np.array(validation_labels)
```

```python
model_predictions = model.predict(validation_data, verbose=1)
```

```python
mp = []
for p in model_predictions:
    if p[0] < p[1]:
        mp.append([0, 1])
    else: 
        mp.append([1, 0])
```

```python
count_based_accuracy = np.equal(validation_labels, mp).sum()/2/len(mp)
print(count_based_accuracy)
```

```python
# count f1
vgg19_f1 = compute_f1(mp, validation_labels.tolist())
```

We see that the precision dropped a bit, while recall improved greatly. Due to the short amount of time used for transfer learning on VGG19, there was only a small increase in the F1 score. Ways to further improve these scores are using a more complicated model, such as ResNet-151, or Inception v3, and of course training longer. Overtraining can cause overfitting.
